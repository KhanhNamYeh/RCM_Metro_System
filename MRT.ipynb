### Union-Find Data Structure

The Union-Find data structure, also known as Disjoint Set Union (DSU), is a data structure that keeps track of a partition of a set into disjoint (non-overlapping) subsets. It provides near-constant-time operations to add new sets, to merge existing sets, and to find the representative member of a set.

#### Key Operations

1. **Initialization**: 
    - `__init__(self, N)`: Initializes the Union-Find structure with `N` elements. Each element is its own set initially.

2. **Find**:
    - `findSet(self, i)`: Finds the representative (or root) of the set containing element `i`. This operation uses path compression to flatten the structure, making future queries faster.

3. **Union**:
    - `unionSet(self, i, j)`: Merges the sets containing elements `i` and `j`. This operation uses union by rank to keep the tree shallow, improving efficiency.

4. **Check if Same Set**:
    - `isSameSet(self, i, j)`: Checks if elements `i` and `j` are in the same set.

### Usage

The Union-Find data structure is particularly useful in network connectivity problems, such as finding connected components in a graph, and in Kruskal's algorithm for finding the Minimum Spanning Tree (MST) of a graph.

```python
class UnionFind:
    def __init__(self, N):
        self.p = list(range(N))
        self.rank = [0] * N
        self.setSize = [1] * N

    def findSet(self, i):
        if self.p[i] == i:
            return i
        else:
            ret = self.findSet(self.p[i])
            self.p[i] = ret
            return ret

    def unionSet(self, i, j):
        if not self.isSameSet(i, j):
            x = self.findSet(i)
            y = self.findSet(j)
            if self.rank[x] > self.rank[y]:
                self.p[y] = x
                self.setSize[x] += self.setSize[y]
            else:
                self.p[x] = y
                self.setSize[y] += self.setSize[x]
                if self.rank[x] == self.rank[y]:
                    self.rank[y] += 1

    def isSameSet(self, i, j):
        return self.findSet(i) == self.findSet(j)
```

### Edge Class and EdgeList Class

The `Edge` class represents an edge in the graph with a weight, source, and destination. It includes methods for comparison and string representation.

#### Edge Class

- `__init__(self, weight, source, dest)`: Initializes an edge with the given weight, source, and destination.
- `__lt__(self, other)`: Compares two edges based on their weight.
- `__str__(self)`: Returns a string representation of the edge.

The `EdgeList` class manages a collection of edges and provides utility functions to add edges, get the size of the edge list, retrieve the list of edges, and print the graph.

#### EdgeList Class

- `__init__(self)`: Initializes an empty edge list.
- `addEdge(self, weight, source, dest)`: Adds an edge to the edge list.
- `getSize(self)`: Returns the number of edges in the edge list.
- `getVector(self)`: Returns the list of edges.
- `printGraph(self)`: Prints all edges in the edge list.

#### Usage

The `EdgeList` class is used to represent the graph of the transportation network. Each edge in the edge list represents a connection between two stations with a certain weight (distance). The edge list is then used in algorithms like Dijkstra's to find the shortest path between stations.

```python
class Edge:
    def __init__(self, weight, source, dest):
        self.weight = weight
        self.source = source
        self.dest = dest

    def __lt__(self, other):
        return self.weight < other.weight

    def __str__(self):
        return f"{self.weight} {self.source} {self.dest}"

class EdgeList:
    def __init__(self):
        self.edges = []

    def addEdge(self, weight, source, dest):
        self.edges.append(Edge(weight, source, dest))

    def getSize(self):
        return len(self.edges)

    def getVector(self):
        return self.edges

    def printGraph(self):
        for edge in self.edges:
            print(edge)

```

### Data Processing: Creating Maps and Coordinates

In this section, we process the data read from the CSV files to create maps and coordinates for the stations. This data is essential for constructing the graph representation of the transportation network and for further analysis and visualization.

#### Creating Maps and Coordinates

- **Map Creation**: 
    - We create a dictionary `map` to store the kind of route and station. Each route is represented as a key, and the value is another dictionary where the keys are station numbers and the values are station IDs.
    - This helps in organizing the stations according to their routes and allows easy access to station IDs based on route and station number.

- **Coordinates Creation**: 
    - We create a dictionary `coornidates` to store the coordinates of each station. The keys are station IDs, and the values are lists containing the latitude and longitude of the stations.
    - This is useful for plotting the stations on a map and for calculating distances between stations.

- **Type Classification**: 
    - We create a dictionary `type` to classify stations into MRT and LRT types. The keys are 'MRT' and 'LRT', and the values are lists of station IDs belonging to each type.
    - This classification helps in distinguishing between different types of stations and can be used for specific analyses or visualizations.

#### Usage

The processed data is used to create the graph representation of the transportation network. The `map` and `coornidates` dictionaries are essential for constructing the edges of the graph, and the `type` dictionary helps in categorizing the stations. This data is then used in algorithms like Dijkstra's to find the shortest path between stations and for visualizing the transportation network.

```python
map = {} 
coornidates = {}  
for s in str_MRT:
    arr = s.strip().split(',')
    coornidate = [float(arr[4]), float(arr[5])]
    coornidates[int(arr[1])] = coornidate
    arr3 = arr[3].split('/')
    for sta in arr3:
        if len(sta) == 3 and sta[2] != 'C':
            routeName = sta[0] + sta[1]
            if routeName not in map:
                map[routeName] = {}
            map[routeName][int(sta[2])] = int(arr[1])
        if len(sta) == 4:
            routeName = sta[0] + sta[1]
            if routeName not in map:
                map[routeName] = {}
            map[routeName][int(sta[2] + sta[3])] = int(arr[1])

type = {'MRT': [], 'LRT': []}
for s in str_MRT:
    arr = s.strip().split(',')
    arr2 = arr[2].split(' ')
    for a in arr2:
        if a == 'MRT':
            type['MRT'].append(int(arr[1]))
        if a == 'LRT':
            type['LRT'].append(int(arr[1]))

print(len(type['MRT']))
print(len(type['LRT']))
```

### Constructing the Edge List

In this section, we construct the edge list for the transportation network. The edge list is created by iterating through the routes and stations, calculating the weights (distances) between consecutive stations, and adding these edges to the `EdgeList` object.

#### Steps

1. **Initialize EdgeList**: 
    - We initialize an empty `EdgeList` object to store the edges of the graph.

2. **Iterate through Routes**: 
    - For each route in the `map` dictionary, we iterate through the stations.

3. **Calculate Weights**: 
    - We calculate the weights (distances) between consecutive stations using the coordinates stored in the `coornidates` dictionary. The weight is calculated as the Euclidean distance between the coordinates of the stations.

4. **Add Edges**: 
    - We add the calculated edges to the `EdgeList` object.

#### Usage

The constructed edge list is used as input for algorithms like Dijkstra's to find the shortest path between stations. The edge list represents the graph of the transportation network, with each edge representing a connection between two stations with a certain weight (distance).

```python
edgeList = EdgeList()
for route in map.keys():
    for station in map[route]:
        max_key = max(map[route].keys())
        previous = 1
        for i in range(2, max_key + 1):
            if i in map[route]:
                w = abs((coornidates[map[route][i]][0] - coornidates[map[route][previous]][0])**2 + (coornidates[map[route][i]][1] - coornidates[map[route][previous]][1])**2)
                edgeList.addEdge(w, map[route][previous], map[route][i])
                previous = i
```

### Dijkstra's Algorithm

Dijkstra's algorithm is a well-known algorithm for finding the shortest paths between nodes in a graph, which may represent, for example, road networks. It was conceived by computer scientist Edsger W. Dijkstra in 1956 and published three years later.

#### Key Concepts

1. **Graph Representation**:
    - The graph is represented using an edge list, where each edge has a source, destination, and weight (distance).

2. **Initialization**:
    - A priority queue (min-heap) is used to efficiently fetch the next node with the smallest tentative distance.
    - Distances from the start node to all other nodes are initialized to infinity, except for the start node itself, which is initialized to zero.

3. **Relaxation**:
    - For each node, the algorithm updates the tentative distances to its neighboring nodes. If a shorter path to a neighbor is found, the distance is updated, and the neighbor is added to the priority queue.

4. **Path Reconstruction**:
    - Once the shortest path to the destination node is found, the path is reconstructed by backtracking from the destination node to the start node using a list of previous nodes.

#### Steps

1. **Initialize**:
    - Set the distance to the start node to 0 and all other nodes to infinity.
    - Push the start node into the priority queue.

2. **Process Nodes**:
    - While the priority queue is not empty, extract the node with the smallest distance.
    - For each neighbor of the extracted node, calculate the tentative distance through the extracted node.
    - If the tentative distance is smaller than the known distance, update the distance and push the neighbor into the priority queue.

3. **Reconstruct Path**:
    - Backtrack from the destination node to the start node using the list of previous nodes to reconstruct the shortest path.

#### Usage

In this project, Dijkstra's algorithm is used to find the shortest path between two stations in a transportation network. The algorithm takes the edge list of the graph, the start station, and the end station as input and returns the shortest distance and the path taken.


```python
def Dijkstra(edgeList, start, end):
    dist = [float('inf')] * (edgeList.getSize() + 1)
    dist[start] = 0
    previous = [None] * (edgeList.getSize() + 1)
    pq = []
    heapq.heappush(pq, (0, start))

    while pq:
        d, u = heapq.heappop(pq)
        if d > dist[u]:
            continue
        for edge in edgeList.getVector():
            if edge.source == u:
                v = edge.dest
                w = edge.weight
                if dist[u] + w < dist[v]:
                    dist[v] = dist[u] + w
                    previous[v] = u
                    heapq.heappush(pq, (dist[v], v))

    # Reconstruct the path from start to end
    path = []
    current = end
    while current is not None:
        path.append(current)
        current = previous[current]
    path.reverse()

    return dist[end], path
```
